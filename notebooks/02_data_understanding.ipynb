## 6. Missing Values Assessment

Missing values reduce the completeness of a dataset and may negatively affect downstream analyses.

This section identifies the number and percentage of missing values present in each variable.

```python
missing_values = df.isnull().sum()

missing_values
```

```python
missing_percentage = (missing_values / len(df)) * 100

missing_percentage.sort_values(ascending=False)
```

### Interpretation

The analysis shows that missing values are primarily concentrated in the **CustomerID** column, with a small number of missing values also present in **Description**.

Missing customer identifiers are particularly important because customer segmentation relies on identifying individual customers.


## 7. Duplicate Records

Duplicate records occur when identical observations appear more than once within the dataset.

These records may distort statistical summaries and sales analyses if left untreated.


```python
duplicates = df.duplicated().sum()
print(f'Duplicate rows: {duplicates}')
```

### Interpretation

The dataset contains duplicate records.

Duplicate transactions may result in the same purchase being counted multiple times, leading to inaccurate calculations of revenue and customer purchasing behaviour.


## 8. Cancelled Transactions


Invoices beginning with the letter **"C"** indicate cancelled customer transactions.

These transactions are identified because they do not represent completed purchases and may influence customer behaviour analysis.

```python
cancelled = df[df["InvoiceNo"].astype(str).str.startswith("C")]

print(f"Cancelled transactions: {len(cancelled)}")

cancelled.head()
```

### Interpretation

The dataset contains cancelled transactions identified by invoice numbers beginning with **"C"**.

These records generally represent product returns or cancelled purchases and should not contribute to analyses of completed customer purchases.

## 9. Negative Quantities

Negative quantities generally represent returned products, stock corrections, or cancelled purchases rather than completed sales.

This section identifies all records containing negative quantities.

```python
negative_quantity = df[df["Quantity"] < 0]

print(f"Negative quantity records: {len(negative_quantity)}")

negative_quantity.head()
```

### Interpretation

Negative quantities are present within the dataset.

These observations typically correspond to returned products or inventory adjustments rather than customer purchases.


## 10. Negative Unit Prices

Unit prices are expected to be positive.

This section investigates whether any transactions contain negative pricing values that may indicate data entry errors or exceptional accounting adjustments.

```python
negative_price = df[df["UnitPrice"] < 0]

print(f"Negative unit prices: {len(negative_price)}")

negative_price.head()
```

### Interpretation

The analysis checks whether any transactions contain negative unit prices.

If such records exist, they may indicate pricing errors or accounting adjustments and should be investigated during the Data Preparation phase.

If no negative prices are found, this suggests that the pricing information is generally consistent.
